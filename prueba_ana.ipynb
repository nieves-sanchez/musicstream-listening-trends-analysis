```python
# Pasos previos al uso del API de Spotify.
# Importamos la librería de Python Spotipy

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import time
```

```python
# Credenciales y proceso de autenticación.
CLIENT_ID = '******'
CLIENT_SECRET = '****'

auth_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
'''Definimos los géneros que queremos estudiar así como el rango de años en el 
que obtener canciones'''
generos = ['rock', 'pop', 'classical', 'reggaeton']
años = range(2019, 2024)

# Lista para almacenar todos los datos
resultados = []
artistas = []

'''Hemos decidido, inicialmente, obtener información sobre 200 canciones por género y año.
Dado que el límite de consulta al API es de 50, debemos configurar la paginación'''      
limite_por_peticion = 50  # Máximo permitido por Spotify
total_canciones_deseadas = 200 # Total que queremos obtener por petición
num_peticiones = total_canciones_deseadas // limite_por_peticion  #4 peticiones    
```

```python
for genero in generos:
    print(f"\nProcesando género: {genero}")
    for año in años:
        print(f"Año: {año}")
        query = f'genre:{genero} year:{año}'
        for peticion in range(num_peticiones):
             offset = peticion * limite_por_peticion
             try:
                 results= sp.search(q=query, type='track', limit=limite_por_peticion, offset=offset, market= 'ES')
                 canciones = results['tracks']['items']
                 if not canciones:
                  print(f"No hay más resultados en offset {offset}")
                  break
                 
                 for item in canciones: 
                  artistas.append(item['artists'][0]['name']) ## añado a la lista de artistas
                  nueva_cancion ={'name_track': item['name'],
                                  'name_track': item['name'],
                                  'type': item['type'], 
                                  'year': item['album']['release_date'],
                                  'genre': genero,
                                  'popularity': item['popularity']}
                  resultados.append(nueva_cancion)
                  
                # Pausa para evitar límites de rate
                 time.sleep(0.5)
             except Exception as e:
              print(f"Error en búsqueda (offset {offset}): {e}")
              break
print(f"Total de canciones extraídas:{len(resultados)}")
    
    
    
    
```

En la celda superior he desarrollado un código para iterar en los diferentes géneros y años, obteniendo resultados salvo para reggaeton. Así mismo, tampoco consigo que en los valores arrojados por resultados aparezcab el género y la popularidad.
A continuación, voy a extraer información género por género iterando únicamente en el rango de años. Creo que así estará más ordenado, guardando cada género en su CSV correspondiente.

```python
#<======= ROCK =========>#
genero = 'rock'
años = range(2019, 2024)
artistas_rock = []
resultados_rock = []

limite_por_peticion = 50  # Máximo permitido por Spotify
total_canciones_deseadas = 200 # Total que queremos obtener por petición
num_peticiones = total_canciones_deseadas // limite_por_peticion  #4 peticiones 

print(f"\nProcesando género: {genero}")
for año in años:
    print(f"Año: {año}")
    query = f'genre:{genero} year:{año}'
    for peticion in range(num_peticiones):  #Paginación
     offset = peticion * limite_por_peticion
     try:
         results_rock= sp.search(q=query, type='track', limit=limite_por_peticion, offset=offset, market= 'ES')
         canciones_rock = results_rock['tracks']['items']
         if not canciones_rock:
          print(f"No hay más resultados en offset {offset}")
          break
         
         for item in canciones_rock: 
          artistas_rock.append(item['artists'][0]['name']) ## añado a la lista de artistas
          nueva_cancion_rock ={'name_track': item['name'],
                'name_artist': item['artists'][0]['name'],
                'type': item['type'], 
                'year': item['album']['release_date'][:4],
                'duration_ms' : item['duration_ms'],
                'genre': genero,
                'popularity': item['popularity']}
          resultados_rock.append(nueva_cancion_rock)
        # Pausa para evitar límites de rate
         time.sleep(0.5)
     except Exception as e:
              print(f"Error en búsqueda (offset {offset}): {e}")
              break
print(f"Total de canciones extraídas:{len(resultados_rock)}")

df_rock = pd.DataFrame(resultados_rock)  
df_rock.to_csv('spotify_data_rock.csv', index=False, encoding='utf-8')
```
